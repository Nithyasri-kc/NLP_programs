**EXPERIMENT - 1 TOKENIZE AND TAG SOME TEXT**

```python
!pip install nltk spacy
import nltk
nltk.download('punkt')
nltk.download('stopwords')
!python -m spacy download en_core_web_sm
def print_text(sample_text, tokens):
    print(f"Original text: {sample_text}")
    print(f"Tokens: {tokens}")
from nltk.tokenize import word_tokenize, sent_tokenize, TweetTokenizer
from nltk.corpus import stopwords
sample_text = "this is a text ready to tokenize"
tokens = word_tokenize(sample_text)
print_text(sample_text, tokens)
tweet_tokenizer = TweetTokenizer()
sample_text = "This is a tweet @jack #NLP"
tokens = tweet_tokenizer.tokenize(sample_text)
print_text(sample_text, tokens)
sample_text = "This is a sentence. This is another one!\nAnd this is the last one."
sentences = sent_tokenize(sample_text)
print_text(sample_text, sentences)
stopwords_ = set(stopwords.words("english"))
sample_text = "this is a sample text"
tokens = sample_text.split()
clean_tokens = [t for t in tokens if not t in stopwords_]
clean_text = " ".join(clean_tokens)
print_text(sample_text, clean_text)
import spacy
nlp = spacy.load("en_core_web_sm")
sample_text = "this is a text ready to tokenize"
doc = nlp(sample_text)
tokens = [token.text for token in doc]
print_text(sample_text, tokens)
sample_text = "This is a sentence. This is another one!\nAnd this is the last one."
doc = nlp(sample_text)
sentences = [sentence.text for sentence in doc.sents]
print_text(sample_text, sentences)
```

**EXPERIMENT - 2 RESUME SCREENING USING NLP**

```python
!pip install spacy
```

```python
!python -m spacy download en_core_web_sm
import spacy
def extract_years_of_experience(doc):
    for ent in doc.ents:
        if ent.text.isdigit() and "year" in ent.text.lower():
            return int(ent.text)
    return 0

def resume_screening(resume_text):
    nlp = spacy.load("en_core_web_sm")
    doc = nlp(resume_text)
    desired_skills = ["python", "machine learning", "data analysis", "communication"]
    extracted_skills = [token.text.lower() for token in doc if token.text.lower() in desired_skills]
    years_of_experience = extract_years_of_experience(doc)
    meets_criteria = (
        "python" in extracted_skills
        and ("machine learning" in extracted_skills or "data analysis" in extracted_skills)
        and years_of_experience >= 2
    )
    print("Extracted Skills:", extracted_skills)
    print("Years of Experience:", years_of_experience)
    print("Meets Screening Criteria:", meets_criteria)

resume_text = """
I am a Python developer with 3 years of experience. My expertise includes machine learning, data
analysis, and effective communication.
"""
resume_screening(resume_text)
```

**EXPERIMENT-3 DEVELOPMENT OF SENTIMENT ANALYSIS WITH PYTHON**

```python
!pip install nltk

import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer

nltk.download('vader_lexicon')
analyzer = SentimentIntensityAnalyzer()
text = "I am going to die."
sentiment_scores = analyzer.polarity_scores(text)
compound_score = sentiment_scores['compound']
if compound_score >= 0.05:
    sentiment = "Positive"
elif compound_score <= -0.05:
    sentiment = "Negative"
else:
    sentiment = "Neutral"
print(f"Sentiment: {sentiment}")
print(f"Compound Score: {compound_score}")
```

**EXPERIMENT - 4 DEVELOP KEYWORD EXTRACTION WITH PYTHON**

```python

!pip install rake-nltk wordcloud matplotlib nltk
import nltk
from rake_nltk import Rake
from wordcloud import WordCloud
import matplotlib.pyplot as plt
nltk.download('punkt')
rake = Rake()
input_text = """
NLP stands for Natural Language Processing.
It is the branch of Artificial Intelligence that gives the ability to machine understand
and process human languages. Human languages can be in the form of text or audio format.
Natural Language Processing started in 1950 When Alan Mathison Turing published
an article in the name Computing Machinery and Intelligence.
It is based on Artificial intelligence. It talks about automatic interpretation and
generation of natural language.
As the technology evolved, different approaches have come to deal with NLP tasks.
"""
rake.extract_keywords_from_text(input_text)
keywords = rake.get_ranked_phrases()
print("Keywords:", keywords)
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(' '.join(keywords))
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

EXP 6 IMPLEMENTING NLP FOR WHATSAPP CHAT

```python

import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.chat.util import Chat, reflections
nltk.download('stopwords')
nltk.download('punkt')
conversation = "Hello! How are you?\nI'm doing well, thanks! What about you?"
tokens = word_tokenize(conversation.lower())
stop_words = set(stopwords.words('english'))
filtered_tokens = [word for word in tokens if word.isalnum() and word not in stop_words]

print("Filtered tokens:", filtered_tokens)
pairs = [
    (r'(hi|hello|hey)', ['Hello!', 'Hi there!', 'Hey!']),
    (r'how are you', ['I am doing well, thank you. How about you?']),
    (r'(.*)', ['I am just a chatbot, not capable of understanding that.']),
]
chatbot = Chat(pairs, reflections)
user_message = "Hi, what can you do?"
response = chatbot.respond(user_message)
print("Chatbot response:", response)
```

EXP 7 CHATBOT IMPLEMENTATION USING NLP

```python

!pip install torch
!pip install transformers
from transformers import GPT2LMHeadModel, GPT2Tokenizer
model_name = "gpt2"
model = GPT2LMHeadModel.from_pretrained(model_name)
tokenizer = GPT2Tokenizer.from_pretrained(model_name)

def chatbot(user_input, model, tokenizer, max_length=100):
    input_ids = tokenizer.encode(user_input, return_tensors="pt")
    output = model.generate(input_ids, max_length=max_length, num_beams=5,
                            no_repeat_ngram_size=2, top_k=50, top_p=0.95)
    response = tokenizer.decode(output[0], skip_special_tokens=True)
    return response

def chat_with_user(model, tokenizer):
    print("Hi, I'm your chatbot. Type 'quit' to exit.")
    while True:
        user_input = input("You: ")
        if user_input.lower() == 'quit':
            print("Goodbye!")
            break
        response = chatbot(user_input, model, tokenizer)
        print("Bot:", response)
if __name__ == "__main__":
    chat_with_user(model, tokenizer)
```

EXP 8 IMPLEMENTATION OF NEXT WORD PREDICTION USING NLP

```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer
import torch
model_name = "gpt2"
model = GPT2LMHeadModel.from_pretrained(model_name)
tokenizer = GPT2Tokenizer.from_pretrained(model_name)
model.eval()
seed_text = "Pizza have different"
next_words = 5
input_ids = tokenizer.encode(seed_text, return_tensors='pt')
output = model.generate(input_ids, max_length=len(input_ids[0]) + next_words,
                        num_return_sequences=1,
                        no_repeat_ngram_size=2,
                        top_k=50,
                        top_p=0.95,
                        temperature=0.7)
generated_text = tokenizer.decode(output[0], skip_special_tokens=True)
print("Next predicted text:", generated_text)
```
